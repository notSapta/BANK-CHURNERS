```python
from sklearn.preprocessing import LabelEncoder                           # Here we Encode 'Gender' as (Male=1, Female=0) and 'Marital_Status' as (0=Divorced, 1=Married, 2=Single, 3=Unknown)
le = LabelEncoder()                                                      # Here we use method LabelEncode() as the variables are nominal(order does not matter)
cols = ['Gender','Marital_Status']
df[cols] = df[cols].apply(LabelEncoder().fit_transform)
df.head()
# type(df)
```

```python
# Here we encode 3 columns : Education_level, Income_Category, Card_Category using ORDINAL ENCODING.
# Unlike Label Encoder, the converted integer value of Ordinal Encoding refers to increasing strength or order.

from sklearn.preprocessing import OrdinalEncoder
Edu = ['Unknown','Uneducated','High School','College','Graduate','Post-Graduate','Doctorate']
Inc = ['Unknown','Less than $40K','$40K - $60K','$60K - $80K','$80K - $120K','$120K +']
Cardc = ['Blue','Silver','Gold','Platinum']
e = OrdinalEncoder(categories=[Edu,Inc,Cardc])
df[['Education_Level','Income_Category','Card_Category']] = e.fit_transform(df[['Education_Level','Income_Category','Card_Category']])
df.head()
```

```python
X = df.iloc[:,1:20].values                 #independent variables
y = df.iloc[:,0].values                    #dependent variable
```

```python
# Splitting the dataset into Training and Testing Parts

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
# Feature Scaling all the variables in X

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# ALGORITHM 1

from sklearn.tree import DecisionTreeClassifier                                 # Implementing DECISION TREE CLASSIFIER to classify
c = DecisionTreeClassifier(criterion='entropy',random_state=0)                  
c.fit(X_train, y_train)                                                         
y_pred = c.predict(X_test)

from sklearn.metrics import confusion_matrix, accuracy_score                    # Importing library to measure the accuracy of our results 
cm = confusion_matrix(y_test, y_pred)                                           # Making Confusion Matrix
ac1 = accuracy_score(y_test, y_pred)                                             # Making Accuracy score.
print('CONFUSION MATRIX :\n',cm)                                                              # Printing CONFUSION MATRIX
print('\nACCURACY : \n:',ac1*100)                                                       # Printing Model ACCURACY
```
